### Data visualizations Assignment 4

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
df.fillna(1)
df['Date'] = pd.to_datetime(df.Date)
```

### Histogram Plot

```python
hist_types = ["Temperature", "Fuel_Price", "Unemployment", "CPI"]

plt.figure(figsize=(16, 10))
i = 0
while i < len(hist_types):
    plt.subplot(2, 2, i+1)
    plt.title('Histogram of {}'.format(hist_types[i]))
    plt.hist(df[hist_types[i]])
    i += 1
```

### Bar Chart

```python
departments = ["dept_1 sales", "dept_2 sales", "dept_3 sales"]

plt.figure(figsize=(20, 40)) 
i = 0

while i < len(departments):
    plt.subplot(4, 1, i+1)
    plt.title('Bar Chart of {}'.format(departments[i]))
    plt.bar(df['Date'], df[hist_types[i]], width = 8)
    i += 1


```

### Line Plot

```python
line_chart_types = ["Temperature", "Fuel_Price", "Unemployment", "CPI"]

plt.figure(figsize=(20, 40)) 

i = 0
while i < len(line_chart_types):
    plt.subplot(4, 1, i+1)
    plt.title('Line Plot of Date v {}'.format(line_chart_types[i]))
    plt.ylabel(line_chart_types[i])
    plt.xlabel('Date')
    df[line_chart_types[i]].plot(y = line_chart_types[i]);
    i += 1  
```

### Scatter Plot

```python
line_chart_types = ["Temperature", "Fuel_Price", "Unemployment", "CPI"]

plt.figure(figsize=(10, 20)) 

i = 0
while i < len(line_chart_types):
    plt.subplot(4, 1, i+1)
    plt.title('Scatter Plot of Date v {}'.format(line_chart_types[i]))
    plt.ylabel(line_chart_types[i])
    plt.xlabel('Date')

    plt.scatter(df['Date'], df[line_chart_types[i]], alpha=0.5)
    i += 1  
    
    
    
```

### Area Plot

```python
departments = ["dept_1 sales", "dept_2 sales", "dept_3 sales"]
plt.figure(figsize=(15, 15)) 

plt.stackplot(df.index, [
                      df['dept_1 sales'], 
                      df['dept_2 sales'],
                      df['dept_3 sales']
                  ],
              labels= departments ,
              alpha=0.8)

plt.legend(loc=2, fontsize='large')
plt.plot(kind='area', stacked=False)
```
