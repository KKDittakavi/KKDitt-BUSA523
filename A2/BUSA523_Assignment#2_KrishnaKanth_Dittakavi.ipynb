Q2. Write a program to take user inputs (number of swords, diamonds, gold coins, ropes and potions) for a video game and store them in a dictionary. After which print the following output.
Inventory:
5 swords
10 diamonds
6 gold coins
3 rope
2 potions

```python
number_of_swords=input("number of swords")
number_of_diamonds=input("diamonds")
number_of_goldcoins=input("goldcoins")
number_of_ropes=input("ropes")
number_of_potions=input("potions")

class my_dictionary(dict):

    # _init_ function
    def _init_(self):
        self = dict()

    # Function to add key:value
    def add(self, key, value):
        self[key] = value


# Main Function
dict_obj = my_dictionary()

dict_obj.add('swords', number_of_swords)
dict_obj.add('diamonds', number_of_diamonds)
dict_obj.add('goldcoins', number_of_goldcoins)
dict_obj.add('ropes', number_of_ropes)
dict_obj.add('potions', number_of_potions)

print(dict_obj)

print("inventory:")
for key in dict_obj:
    print(dict_obj.get(key), key)
```

Q4.. Create a game with the following instructions:
a. There are 3 players and 10 iterations.
b. In each iteration, every player rolls a die.
c. The winner of the game is the one who has the highest score when rollsfrom all the iterations are added.

```python
import random


def roll():
    return random.randint(1,6)


def player_dice(iteration=1):
    total_player_1 = 0
    total_player_2 = 0
    total_player_3 = 0

    for i in range(1, iteration + 1):
        total_player_1 += roll()
        #print(total_player_1)
        total_player_2 += roll()
        #print(total_player_2)
        total_player_3 += roll()
        #print(total_player_3)
        print("p1", total_player_1, "p2", total_player_2,"p3",total_player_3)
        
    if total_player_1 > total_player_2 and total_player_2 > total_player_3:
        print("Player 1 WINS!!", total_player_1)
    elif total_player_2 > total_player_1 and total_player_2 > total_player_3:
        print("Player 2 WINS!!", total_player_2)
    elif total_player_3 > total_player_2 and total_player_2 > total_player_1:
        print("Player 3 WINS!!", total_player_3)

    return

player_dice(10)
```
